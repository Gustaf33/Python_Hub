```python
#example of a dataset from scikit-learn

import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
#Imports for plotting

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
#Now import dataset from scikit learn as well as the linear_model module. 
#Note: You may have to run a download, scikit learn will read an error and prompt you to if you don't have the dataset

from sklearn.datasets import fetch_california_housing
from sklearn.datasets import fetch_openml
```

```python
# Load the housing dataset
housing = fetch_openml(name="house_prices", as_frame=True)
```

```python
#Let's see what the data set contains
print(housing.DESCR)
```

```python
# Histogram of prices (this is the target of our dataset)
plt.hist(housing.target,bins=50)

#label
plt.xlabel('Price in $1000s')
plt.ylabel('Number of houses')
```

```python
# Plot the column at the 5 index (Labeled RM)
plt.scatter(housing.data[:,5],housing.target)

#label
plt.ylabel('Price in $1000s')
plt.xlabel('Number of rooms')
```

```python
# reset data as pandas DataFrame
House_df = DataFrame(housing.data)

# label columns
House_df.columns = housing.feature_names

#show
House_df.head()
```
